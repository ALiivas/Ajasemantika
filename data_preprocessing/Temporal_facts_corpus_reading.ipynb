### Saving Text-objects in JSON-format

```python
# -- new target directory for temporal facts corpus articles in JSON format
json_dir = 'temporal_facts_corpus_json'

if not os.path.isdir(json_dir):
    os.mkdir(json_dir)

# -- target folder path
json_dir_path = 'temporal_facts_corpus_json/'

n_saved_articles = 0
for text_obj in text_objects:
    print(text_obj.meta['filename'])
    
    # save Text-object in JSON format
    save_Text_to_json(json_dir_path, text_obj)
    n_saved_articles+=1
    
print()
print(f"Saved {n_saved_articles} articles in JSON-format.")
```
