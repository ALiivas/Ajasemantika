### Running pipeline on all EstTimeML corpus articles

```python
baseAnnotations = load_base_segmentation(baseAnnotationFile)
(eventsByLoc, eventsByID) = load_entity_annotation(eventAnnotationFile)
articlesDCT = load_articles_DCT(articleMetadata_file)
(timexesByLoc, timexesByID) = load_entity_annotation(timexAnnotationFile)
```

```python
n_saved_articles = 0
for filename in articlesDCT.keys():
    print(filename)
    
    article_text, mapping = create_Text_object(filename, articlesDCT, baseAnnotations, eventsByLoc, timexesByLoc)
    
    gold_word_events = create_gold_word_events_layer(article_text, baseAnnotations.get(filename))
    gold_event_phrases = create_gold_event_phrases(article_text)
    gold_timex_phrases = create_gold_timex_phrases(article_text)
    
    article_text = add_layers_to_text(filename, article_text, articlesDCT, [gold_word_events, gold_event_phrases, gold_timex_phrases])
    
    # Text-object is saved to JSON file
    save_Text_to_json(article_text)
    n_saved_articles+=1
    
print()
print("Saved", n_saved_articles, "articles in JSON-format.")
```
