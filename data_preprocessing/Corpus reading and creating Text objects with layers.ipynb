```python
def create_event_timex_rel_layer(article_text, relations):
    tlinks_layer = RelationLayer('event_timex_tlinks', span_names=['a_text', 'b_text'], 
                                 attributes=['a_ID', 'rel_type', 'b_ID', 'comment'], 
                                 display_order=['a_text', 'a_ID', 'rel_type', 'b_text', 'b_ID', 'comment'], 
                                 text_object=article_text, enveloping='words')
    
    last_ann = None
    
    for entity in relations:
        ann = relations.get(entity)[0]
        
        if last_ann==ann:
            last_ann == None
            continue 
            
        else:
            last_ann = ann
            a_ID = ann[0]
            rel_type = ann[1]
            b_ID = ann[2]
            comment = ann[3]
            
            a_event = [e for e in article_text.gold_event_phrases if e['event_ID'][0]==a_ID]
            b_event = [e for e in article_text.gold_event_phrases if e['event_ID'][0]==b_ID]
            
            if a_event:
                b_timex = [t for t in article_text.gold_timex_phrases if t['timex_ID'][0]==b_ID]
                a_spans = [span.base_span for span in a_event[0]]
                b_spans = [span.base_span for span in b_timex[0]]
                
                tlinks_layer.add_annotation(a_text=a_spans, a_ID=a_ID, rel_type=rel_type,
                                            b_text=b_spans, b_ID=b_ID, comment=comment)
                
            elif b_event:
                a_timex = [t for t in article_text.gold_timex_phrases if t['timex_ID'][0]==a_ID]
                a_spans = [span.base_span for span in a_timex[0]]
                b_spans = [span.base_span for span in b_event[0]]
                
                tlinks_layer.add_annotation(a_text=a_spans, a_ID=a_ID, rel_type=rel_type,
                                            b_text=b_spans, b_ID=b_ID, comment=comment)
                
    return tlinks_layer
```

```python
tlinks = create_event_timex_rel_layer(article_text, event_timex_relations.get('aja_ml_2002_47.tasak.a023.sol'))
```

```python
article_text = add_layers_to_text('aja_ml_2002_47.tasak.a023.sol', article_text, articlesDCT, 
                                  [tlinks])
```

### Running pipeline on all EstTimeML corpus articles

```python
baseAnnotations = load_base_segmentation(baseAnnotationFile)
(eventsByLoc, eventsByID) = load_entity_annotation(eventAnnotationFile)
articlesDCT = load_articles_DCT(articleMetadata_file)
(timexesByLoc, timexesByID) = load_entity_annotation(timexAnnotationFile)
```

```python
n_saved_articles = 0
for filename in articlesDCT.keys():
    print(filename)
    
    article_text, mapping = create_Text_object(filename, articlesDCT, baseAnnotations, eventsByLoc, timexesByLoc)
    
    gold_word_events = create_gold_word_events_layer(article_text, baseAnnotations.get(filename))
    gold_event_phrases = create_gold_event_phrases(article_text)
    gold_timex_phrases = create_gold_timex_phrases(article_text)
    
    article_text = add_layers_to_text(filename, article_text, articlesDCT, [gold_word_events, gold_event_phrases, gold_timex_phrases])
    
    # Text-object is saved to JSON file
    save_Text_to_json(article_text)
    n_saved_articles+=1
    
print()
print("Saved", n_saved_articles, "articles in JSON-format.")
```
